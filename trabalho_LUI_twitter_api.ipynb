```python
df = pd.DataFrame(columns=['created_at', 'text', 'hashtag1', 'hashtag2', 'hashtag3', 'user_mentions1','user_mentions2','user_mentions3'])
```

```python
df
```

```python
count = 0
for tweet in tweepy.Cursor(api.search, q=query, count=qtd_tweet, lang=lang, since="2021-01-29").items(qtd_tweet):

  hashtags_list = ["","",""]
  user_mentions_list = ["","",""]

  if len(tweet.entities["hashtags"]) != 0:
    for i, item in enumerate(tweet.entities["hashtags"]):
      hashtags_list[i] = item["text"]
      if i == 2:
        break
  
  if len(tweet.entities["user_mentions"]) != 0:
    for i, item in enumerate(tweet.entities["user_mentions"]):
      user_mentions_list[i] = item["screen_name"]
      if i == 2:
        break

  df.loc[count] = [tweet.created_at]+[tweet.text]+[hashtags_list[0]]+[hashtags_list[1]]+[hashtags_list[2]]+[user_mentions_list[0]]+[user_mentions_list[1]]+[user_mentions_list[2]]
  
  count += 1
    
```
